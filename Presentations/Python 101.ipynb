# Hands-on Exercise

Create a class named Rectangle, which requires height and width as parameters. 
Create methods area and perimeter that calculater area and perimeter of the rectangle with the provided height and width.

## Solution

```python
class Rectangle:
    def __init__(self, height, width):
        self.height = height
        self.width = width
        
    def area(self):
        return self.height * self.width
    
    def perimeter(self):
        return 2 * (self.height + self.width)
```

```python
r = Rectangle(5 , 10)
```

```python
r.area()
```

```python
r.perimeter()
```
